```python
import pandas as pd
import numpy as np
url="https://raw.githubusercontent.com/Sankari-S1836/food-waste-ml/refs/heads/main/food_waste_data_sample.csv"
d=pd.read_csv(url)
d
```

```python
d.isnull().sum()
```

```python
d.drop(['expiry','name'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
d['category']=le.fit_transform(d['category'])

d
#in is_wasted=1()  whether a food item is likely to be wasted based on its quantity and how soon it will expire.

```

```python
import seaborn as sns
sns.heatmap(d.corr())
```

```python
x=d.drop(columns=['is_wasted'])
y=d['is_wasted']
```

```python
# training model
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import r2_score
ypred=lr.predict(xtest)
r2_score(ytest,ypred)
```
